# Visualization

### Let's compare the countries of Africa based on the number of conflicts that happened from 2007 to 2016. What's the country with the most conflicts?And the least? Do neighbouring countries share the same conflict trend? - This interavtive map will answer the above questions. 

```python
conf_map = alt.Chart(africa_p, title='Total number of conflicts from 1997 to 2016 by country: Africa').configure_title(fontSize=24).transform_lookup(
    lookup = 'admin',
    from_ = alt.LookupData(data = sum_conf, 
                           key = 'country',
                           fields=['size', 'sum', 'mean'])
).mark_geoshape(color ='lightgray', stroke = 'white', strokeWidth = 1
).encode(
    color = alt.Color('size:Q', title='Number of conflicts:',scale=alt.Scale(scheme='reds', reverse = False)),    
) 

conf_map.encode(tooltip = [
        alt.Tooltip('admin:N', title='Country'),
        alt.Tooltip('size:Q', title='Number of conflicts'),
        alt.Tooltip('sum:Q',title='Number of fatalities'),
        alt.Tooltip('mean:Q', title='Average number of fatalities', format=",.2f")
        ],
               
        opacity = alt.condition(highlight, 
                        alt.value(0.5), alt.value(1)),
        strokeWidth= alt.condition(highlight, 
                           alt.value(3), alt.value(0.5))).add_selection(highlight).properties(width = 800, height = 600).configure_view(strokeWidth = 0)
```

#### Pros: I have used this map over other charts for this task to illustrate the geographical location of the numbers - we can see the geo trends of conflicts. The color palette 'reds' was best suite, as conflicts is smth to be negalivety percepted. Also added Tooltips for more precise info on the number of conflicts and number of fatalities.

#### Cons: While being a good illustration of geographic location of conflicts, the coding the number of conflicts with this scale might be not most illustratiive. There are a few regions with similar colors - so it might be hard to match the slight difference in colors to difference in numbers. Also, even though the Tooltip helps to be more precise, user should memorize the number of conflicts for one country and then hower to a different one to be able to compare two different countries. 

#### Overall, for the questions listed above, an interactive map is a good choice.  

#see the distribution of conflict reasons by country

```python
count_et = all_dt.groupby(['COUNTRY', 'EVENT_TYPE'], as_index=False)['EVENT_TYPE'].agg(['count'])
count_et['country'] = count_et.index.get_level_values(0)
count_et['EVENT_TYPE'] = count_et.index.get_level_values(1)
count_et.reset_index(drop=True, inplace=True)
count_et
```

## What are the types of conflicts? Which ones are the most frequent? Let's see what is the conflict picture for each of the countries. Use a selector on the bottom of the chart to see information on a specific country

```python

input_dropdown = alt.binding_select(options=sorted(list(set(count_et['country'].values))), name='Country')
selection = alt.selection_single(fields=['country'], bind=input_dropdown, init={"country":"Botswana"})


alt.Chart(count_et, title='Total number of conflicts by type for a selected country').configure_title(fontSize=24).configure_axis(
    labelFontSize=13, titleFontSize=15
).mark_bar(color='maroon').encode(
    x=alt.X('count:Q', title='Total number of conflicts from 2007 to 2016'),
    y=alt.Y('EVENT_TYPE:N', sort='-x', title=None),
    tooltip = [
        alt.Tooltip('count:Q', title='Number of conflicts')]
    
).add_selection(selection).transform_filter(selection).properties(width = 800, height = 300)
```

#### I chose this type of chart for few reasons: it shows the conflict type break-down structure and it is easy-to-read. Initially, I was going to plot conflict break-down only for one country, but the the choice of it would have been hardcoded. Adding a selector for country name lets the user see the frequency of conflict types per any country. I also added alphabetical sort for the dropdown - as it is much easier to find the desired country

#### Pros: It is really clear what type of conflicts were present in a specific country and in which volume(frequency).  No extra elements, answers on the stated questions above. User can select any African country from the data available. Can be combined with the first visualization. 

#### Cons: User needs to select the country manually. The dropdown could have been configured better (fe placed near the chart title so that it replaces 'selected country'). Also, it is not suitable for comparing 2 countries - only for analyzing the conflict types within the country.

#### Overall, simple chart for a simple question. 

```python
conf_year = all_dt.groupby(['COUNTRY', 'YEAR'], as_index=False)['COUNTRY'].agg(['count'])
conf_year['country'] = conf_year.index.get_level_values(0)
conf_year['year'] = conf_year.index.get_level_values(1)
conf_year.reset_index(drop=True, inplace=True)
conf_year
```

### Now, what if we want to see how the conflict number has changed over time? Was Somalia always such conflicting? How did conflict number change for each country over the years? - Using faceting we can see an area chart for every country to see the general trend for the country. For comparison, I added the tooltip with mean value which are more representative. Also, the plots are displayed in descending order relative to the mean conflicts frequency. I left all the axis labeled, because I think it helps to read the plot more easily - user doesn't need to memorize what was the axis names and meaning.  

```python
highlight = alt.selection_single(on='mouseover', nearest=False, fields=['country'], empty='none') # add a single selection to highlight chart objects when shoving the mouse over 

linear = alt.Chart(conf_year).mark_area(color='maroon').encode(
    x = alt.X('year:Q', title="year"),
    y = alt.Y('count:Q', title="Conflicts"),
    
    opacity = alt.condition(highlight, 
                            alt.value(1), alt.value(0.7)),
    tooltip = [alt.Tooltip('country:N', title='Country:'),
              alt.Tooltip('mean(count):N', title='Mean:'),
              alt.Tooltip('year:N', title='Year:')]


).add_selection(highlight

).properties(width = 220, height = 110).facet(facet=alt.Facet('country:N', sort=alt.Sort(field='count', op='mean', order='descending'),header=alt.Header(labelFontSize=13), title=None), columns=4).resolve_scale(
    x='independent', 
    y='independent'
).configure_axis(
    labelFontSize=11, titleFontSize=11
)


linear.properties(title=alt.TitleParams(text='Number of conflicts in Africa increased in majority of countries over years', 
                                        subtitle='Conflict number per country from 2007 to 2016'), padding=25
                 ).configure_title(fontSize=24, subtitleFontSize=18)
```
